```python
import pandas as pd
```

```python
data = {
    'A': [1, 2, None, 4, 5],
    'B': [None, 2, 3, 4, 5],
    'C': [1, 2, 3, None, 5]
}
df = pd.DataFrame(data)
```

```python
# Dropping rows

df_dropna = df.dropna()
print("After dropping rows with missing values:\n", df_dropna)
```

```python
data_2 = {
    'A': [1, 2, None, 4, 5],
    'B': [None, 2, 3, 4, 5],
    'C': [1, 2, 3, None, 5]
}
df_2 = pd.DataFrame(data)
```

```python
# Replacing with mean/median/mode
df_mean = df_2.fillna(df_2.mean())
print("After replacing missing values with mean:\n", df_mean)

df_median = df_2.fillna(df_2.median())
print("After replacing missing values with median:\n", df_median)

df_mode = df_2.fillna(df_2.mode().iloc[0])
print("After replacing missing values with mode:\n", df_mode)
```
